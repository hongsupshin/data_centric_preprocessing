```python
example_obj = pd.DataFrame(
    {
        'nan': [np.nan, np.nan],
        'nan_none': [np.nan, None],
        'nan_array': [np.nan, []],
        'nan_bool': [np.nan, True],
        'nan_float': [np.nan, 1.0],        
        'nan_int': [np.nan, 1],
        'none': [None, None],        
        'none_array': [None, []],
        'none_bool': [None, True],
        'none_float': [None, 0.0],
        'none_int': [None, 1],
        'array': [[], []],
        'array_bool': [[], True],
        'array_float': [[], 1.0],
        'array_int': [[], 1],
        'bool': [True, False],
        'bool_float': [True, 0.0],
        'bool_int': [True, 1],
        'float_int': [1.0, 0],
    },
    dtype=object
)

example_obj_results = example_obj.T
example_obj_results['pandas'] = example_obj.infer_objects().dtypes
example_obj_results['pandas_api'] = example_obj.apply(lambda x:pd.api.types.infer_dtype(x, skipna=True))
example_obj_results = example_obj_results.T

example_obj_results
```
